Consultas a la API del firewall
===========================================

Se utiliza la biblioteca Requests para hacer la conexión con el firewall.
La primera petición es para autenticar al usuario y poder realizar consultas y/o cambios.

```python
import requests
from requests.auth import HTTPDigestAuth
url = "https://nsa6700.demo.sonicwall.com/api/sonicos/auth"
url2 = "https://nsa6700.demo.sonicwall.com/api/sonicos/address-objects/ipv4"
encabezados = {
   "Accept" : "application/json",
   "Content-type" : "application/json"
}
data = {
   "override" : "true"
}

r = requests.post(url=url, auth=HTTPDigestAuth('demo', 'password'), headers=encabezados, json=data)
```

La autenticación fue exitosa si recibimos el siguiente código ([HTTP status code](http://en.wikipedia.org/wiki/List_of_HTTP_status_codes) 200)

```python
r.status_code
```

Ahora se hará la petición para traer la información de los objetos tipo "Hosts"

```python
r2 = requests.get(url2, headers=encabezados)
```

Nuevamente, un valor 200 indica que la petición fue exitosa:

```python
r2.status_code
```

Con el siguiente comando podemos ver los objectos en formato JSON

```python
r2.json()
```

Por último, cerramos la sesión del usuario:

```python
r3 = requests.delete(url, headers=encabezados)
r3.status_code
```
